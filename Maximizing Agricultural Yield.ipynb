#### Dataset Overview
The dataset comprises 2200 entries with the following columns:
- **Nitrogen**: Nitrogen level in the soil
- **Phosphorus**: Phosphorus level in the soil
- **Potassium**: Potassium level in the soil
- **Temperature**: Temperature in Celsius
- **Humidity**: Humidity percentage
- **pH_Value**: pH value of the soil
- **Rainfall**: Rainfall in mm
- **Crop**: Type of crop

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("/kaggle/input/crop-recommendation-dataset/Crop_Recommendation.csv")
```

```python
# 1. Distribution of Nitrogen levels across different crops
plt.figure(figsize=(12, 6))
sns.boxplot(x='Crop', y='Nitrogen', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Nitrogen Levels Across Different Crops')
plt.show()
```

```python
# 2. Scatter plot of Temperature vs. Humidity colored by Crop
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Temperature', y='Humidity', hue='Crop', data=df, palette='tab20')
plt.title('Temperature vs. Humidity Colored by Crop')
plt.show()
```

```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# 3. Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# 4. Pair plot of all numerical features colored by Crop
sns.pairplot(df, hue='Crop', palette='tab20', plot_kws={'alpha': 0.5})
plt.suptitle('Pair Plot of All Numerical Features Colored by Crop', y=1.02)
plt.show()
```

```python
# 5. Distribution of pH values across different crops
plt.figure(figsize=(12, 6))
sns.violinplot(x='Crop', y='pH_Value', data=df, inner='quartile')
plt.xticks(rotation=90)
plt.title('Distribution of pH Values Across Different Crops')
plt.show()
```

```python
# 6. Bar plot of average nutrient levels (N, P, K) for each crop
plt.figure(figsize=(14, 7))
nutrients = df.groupby('Crop')[['Nitrogen', 'Phosphorus', 'Potassium']].mean().reset_index()
nutrients = pd.melt(nutrients, id_vars='Crop', value_vars=['Nitrogen', 'Phosphorus', 'Potassium'], var_name='Nutrient', value_name='Level')
sns.barplot(x='Crop', y='Level', hue='Nutrient', data=nutrients)
plt.xticks(rotation=90)
plt.title('Average Nutrient Levels for Each Crop')
plt.show()
```

```python
# 7. Box plot of Rainfall for different crops
plt.figure(figsize=(12, 6))
sns.boxplot(x='Crop', y='Rainfall', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Rainfall Across Different Crops')
plt.show()
```

```python
# 8. Histogram of Temperature distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Temperature'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# 9. Line plot of average Temperature and Humidity for each crop
avg_temp_humidity = df.groupby('Crop')[['Temperature', 'Humidity']].mean().reset_index()
plt.figure(figsize=(14, 7))
sns.lineplot(x='Crop', y='Temperature', data=avg_temp_humidity, marker='o', label='Temperature')
sns.lineplot(x='Crop', y='Humidity', data=avg_temp_humidity, marker='o', label='Humidity')
plt.xticks(rotation=90)
plt.title('Average Temperature and Humidity for Each Crop')
plt.ylabel('Level')
plt.legend()
plt.show()
```

```python
# 10. Swarm plot of Potassium levels across different crops
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Crop', y='Potassium', data=df)
plt.xticks(rotation=90)
plt.title('Swarm Plot of Potassium Levels Across Different Crops')
plt.show()
```
